```python
from datetime import date

data = response.json()
products = data.get("products", [])

snapshot_date = date.today().isoformat()
extracted_items = []

#loop through the products and extract features
for p in products:
    if not isinstance(p, dict):
        continue
    ALLOWED = {
        "Crazy_Deal",
        "Festive_Price_Crash",
        "GST_Benefit_Included",
        "Myntra_Unique"
        }

    tag_groups = p.get("styleTagEntry", {}).get("tagGroups", {})

    promotion_tags = [
        t["tagId"]
        for block in ("pricing", "selection")
        for t in tag_groups.get(block, {}).get("tagInfos", [])
        if t.get("tagId") in ALLOWED
        ]
    extracted_items.append({
        "product_id": p.get("productId"),
        "brand": p.get("brand"),
        "gender": p.get('gender'),
        "category": p.get("category"),
        "master_category": (p.get("masterCategory") or {}).get('typeName'),
        "sub_category": (p.get('subCategory') or {}).get('typeName'),
        "article_type": (p.get("articleType") or {}).get('typeName'),
        "mrp": p.get("mrp"),
        "price": p.get("price"),
        "discount": p.get("discount"),
        "discount_type": p.get("discountType"),
        "discount_display_label": p.get("discountDisplayLabel"),
        #"coupon_data" = p.get("couponData") or {},
        "coupon_discount" : (p.get("couponData") or {}).get("couponDiscount"),
        "best_price": ((p.get("couponData") or {}).get("couponDescription") or {}).get("bestPrice"),
        "rating": p.get("rating"),
        "rating_count": p.get("ratingCount"),
        "list_views": p.get("listViews"),
        "inventory": sum(i.get("inventory", 0) for i in p.get("inventoryInfo", [])),
        "available": any(i.get("available", False) for i in p.get("inventoryInfo", [])),
        "sizes": p.get("sizes"),
        "has_multiple_sizes": p.get("hasMultipleSizes"),
        "snapshot_date": snapshot_date,
        "season": p.get("season"),
        "is_fast_fashion": p.get("isFastFashion"),
        "promotion_tags": "|".join(promotion_tags),
        "has_promotion": bool(promotion_tags),
        "year": p.get('year'),
        "preferred_delivery_tag": p.get("preferredDeliveryTag"),
        "delivery_promise": p.get("deliveryPromise")
    })

```

```python

dff = pd.DataFrame(extracted_items)
print(dff.shape)
dff.tail()
```
